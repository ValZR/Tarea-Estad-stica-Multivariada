# Implementar LDA

Creamos un conjunto de datos

```python
import numpy as np
#Primero necesitamos un vector de medias
#Lo ponemos centrado
mu1=[0,0]
#matriz de varianzas
#el array transforma una lista de listas
#necesitamos una matriz positiva definida
sigma1= np.array([[1,0.5],[0.5,1]])
#creamos otra matriz
sigma2=np.array([[1,0.5],[0.5,1]])
#necesitamos el conjunto de datos
#media miu 1 sigma 1, matriz de covarianzas sigma 1
X=np.random.multivariate_normal(mu1, sigma1,5000)
X
#Estos puntos son puntos aleatorios que tienen una distribucion normal ultivariada con matriz de covarianzas 2x2
```

```python
#La hipotesis importante para el LDA
#tenemos una frontera de decision entre dos fronteras de densidad
#la hipotesis para tener linealidad es la homogeneidad
#ahora vamos a generar otra distribuci√≥n que va a tener el mismo vector de medias
#y tambien va a tener la misma matriz de covarianza y con eso vamos a poder hacer una frontera
#es el cociente de las pis
#la frontera de decision es lineal
#decidir a traves de una regla bayesiana que punto se va a ir a que grupo
#del conjunto de datos que generamos vamos a tomar 2000
#estamos reduciendo el conjunto de datos
#cuando tenemos un arreglo y tenemos los corchetes se les llaman accesos porque es acceder a datos de la matriz
#lo primero que se hace es tomar las primeras 2000 entradas de el primer eje en esa columna
#se les suma 2 para que los numeros sean positivos
#la visualizacion de los datos se va a ver mejor
X[2000:,0]=X[2000:,0]+2
X[2000:,1]=X[2000:,1]+2
```

```python
#vamos a hacer un flattening
#vector de 5000x1
y=np.ones((5000,))
#a partir de este elemento va a poner ceros
y[:2000]=0
```

```python
#escojo una matriz de 6x2
#los dos puntos son secuencias
X[:6,:]
```
